```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Creating a FacetGrid to compare total_bill vs tip for each day and smoker status

sns.FacetGrid(data = tips, col = 'day', row  = 'smoker').map(sns.scatterplot, 'total_bill', 'tip')
plt.suptitle("FacetGrid of Total Bill vs Tip by Day and Smoker Status", y = 1.02)
plt.savefig('003.FacetGrid_Total_Bill_vs_Tip.png')
```
