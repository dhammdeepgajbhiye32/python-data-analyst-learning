```python
# Question 2: Build a clustered heatmap (clustermap) using correlation values from the tips dataset
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# using clustermap to visualize the correlation values of the tips dataset

column = ['total_bill','tip','size']
sns.clustermap(tips[column].corr(), annot = True, cmap = 'coolwarm')
plt.title("Clustered Heatmap of Correlation Values")
plt.savefig('002.Clustered_Heatmap_Correlation_Values.png')
```
