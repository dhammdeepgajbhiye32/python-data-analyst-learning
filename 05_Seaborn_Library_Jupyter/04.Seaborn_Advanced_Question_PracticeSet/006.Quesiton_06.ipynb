```python
# Question 6: Build a multi-variable catplot comparing tip, day, sex and smoker.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Building a multi-variable catplot comparing

sns.catplot(data = tips, x = 'day', y = 'tip', hue = 'sex', col = 'smoker', kind = 'bar')
plt.suptitle("Multi-variable Catplot Comparing Tip, Day, Sex, and Smoker")
plt.savefig('006.Multi_variable_Catplot_Tip_Day_Sex_Smoker.png')
```
