```python
# Question 8: Use PairGrid to create: scatter plots in upper triangle, histograms on diagonal, KDE plots in lower triangle
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loadint the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Using PairGrid to create the required plots

g = sns.PairGrid(data = tips, hue = 'sex')
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.add_legend()
plt.suptitle('PairGrid with scatter plots, histograms and KDE plots', y = 1.02)
plt.savefig('008.PairGrid_with_scatter_histograms_KDE_plots.png')
```
