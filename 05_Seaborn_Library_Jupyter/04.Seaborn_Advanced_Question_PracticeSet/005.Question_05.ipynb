```python
# Question 5: Create a regression plot showing the trend between total_bill and tip
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Creating a regression plot to show the trend between total_bill and tip

sns.regplot(data = tips, x = 'total_bill', y = 'tip', color = 'green')
plt.title("Regression Plot of Total Bill vs Tip")
plt.savefig('005.Regression_Plot_Total_Bill_vs_Tip.png')
```
