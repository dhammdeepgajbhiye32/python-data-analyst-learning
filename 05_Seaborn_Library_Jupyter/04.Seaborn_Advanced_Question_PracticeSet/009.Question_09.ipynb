```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loadint the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Creating a heatmap with annotations showing correlation values

from pyparsing import line


columns = ['total_bill', 'tip', 'size']
correlation_matrix = tips[columns].corr()

sns.heatmap(correlation_matrix, annot = True, linewidths= 2, linecolor = 'black', cmap = 'Blues')
plt.title('Correlation Heatmap with Annotations')
plt.savefig('009.Heatmap_with_annotations_correlation_values.png')
```
