```python
# Question 1: Create a pariplot where points are colored by 'sex'.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset 
tips = sns.load_dataset('tips')
tips
```

```python
# Creating a pairplot where points are colored by 'sex'
sns.pairplot(data = tips,vars= ['total_bill', 'tip', 'size'], hue = 'sex')
plt.title("PairPlot Colored with Total_bill, Tip and Size")
plt.savefig('001.Pairplot_Colored_by_Sex.png')
```
