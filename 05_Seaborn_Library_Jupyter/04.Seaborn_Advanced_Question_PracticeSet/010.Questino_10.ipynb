```python
# Question 10: Build a complete dashboard-style visualization combining 1. barplot, 2. boxplot, 3. scatter plot and 4. heatmap in a single figure using subplots
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loadint the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Creating a dashboard-style visualization with subplots

fig, axes = plt.subplots(2,2, figsize = (10,10))

# 1. Barplot

sns.barplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', ax = axes[0,0])
axes[0,0].set_title('Barplot of Total_bill by Day and Sex')

# 2. Boxplot

sns.boxplot(data = tips , x = 'day', y = 'total_bill', hue = 'sex', ax = axes[0,1])
axes[0,1].set_title('Boxplot of Total_bill by Day and Sex')

# 3. Scatter plot

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', ax = axes[1,0])
axes[1,0].set_title('Scatter plot of Total_bill vs Tip by Sex')

# 4. Heatmap of correlation

columns = ['total_bill','tip','size']
correlation_matrix = tips[columns].corr()
sns.heatmap(data = correlation_matrix, annot = True, cmap = 'coolwarm', ax = axes[1,1])
axes[1,1].set_title('Heatmap of Correlation between Numerical Variables')

plt.suptitle('Dashboard-style Visualization with Barplot, Boxplot, Scatter plot and Heatmap', y = 1.02)
plt.tight_layout()
plt.show()

plt.savefig('010.Dashboard_containing_4_different_plots.png')
```
