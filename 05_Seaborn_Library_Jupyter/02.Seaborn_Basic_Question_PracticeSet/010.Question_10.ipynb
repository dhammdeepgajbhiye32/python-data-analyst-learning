```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# changing the theme style to darkgrid
sns.set_style('darkgrid')
sns.set_context('talk')
# Recreating a scatter plot of total_bill vs tip with the new theme
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'time', palette = 'Set2')
plt.title("Scatter Plot of Total Bill vs Tip with Darkgrid Theme")
plt.savefig('010.Scatter_Plot_with_Darkgrid_Theme.png')
```
