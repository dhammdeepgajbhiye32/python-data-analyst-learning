```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# Creating barplot showing the average tip for each day

sns.barplot(data = tips, x = 'day', y = 'tip', palette = 'Set2')
plt.title("Barplot of Average Tip for Each Day")
plt.savefig('006.Barplot_of_Average_Tip_For_Each_Day.png')
```
