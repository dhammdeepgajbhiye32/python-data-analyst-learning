```python
# Question 9: Use pairplot() to visualize the relationships between all numberical variables in the tips dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
columns = ['total_bill', 'tip', 'size']

sns.pairplot(tips[columns], diag_kind = 'kde', palette = 'Set2')
plt.suptitle("Pairplot of Numerical Variables in the Tips Dataset", y = 1.02)
plt.savefig('009.Pairplot_of_Numerical_Variables_in_Tips_Dataset.png')
```
