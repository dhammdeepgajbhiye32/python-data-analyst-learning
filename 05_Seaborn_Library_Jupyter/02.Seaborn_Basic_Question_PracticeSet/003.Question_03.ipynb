```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.histplot(data = tips, x = 'total_bill', kde = True)
plt.title("Histogram of Total Bill")
plt.savefig('003.Histogram_of_Total_Bill.png')
```
