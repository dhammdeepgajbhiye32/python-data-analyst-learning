```python
# Question 7: Use hue to visualize how smoker vs non-smoker affects the tips in a barplot.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# Creating a barplot to visualize how smoker vs non-smoker affects the tips

sns.barplot(data = tips, x = 'day', y = 'tip', hue = 'smoker', palette = 'Set2')
plt.title("Barplot of Tips by Day and Smoker Status")
plt.savefig('007.Barplot_of_Tips_by_Day_and_Smoker_Status.png')
```
