```python
# Question 5: Plot a countplot to show how many customers visited on each day.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Creating a countplot to show how many customers visited on each dat.

sns.countplot(data = tips, x = 'day', palette = 'Set2')
plt.title("Countplot of Customers Visited on Each Day")
plt.savefig('005.Countplot_of_Customers_Visited_on_Each_Day.png')
```
