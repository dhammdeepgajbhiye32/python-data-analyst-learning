```python
# Question 8: Plot a violin plot to show the distribution of tip across time (Lunch vs Dinner).
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# Creating a violin plot to show the distribution of tip across time (Lunch vs Dinner)

sns.violinplot(data = tips, x = 'time', y = 'tip', palette = "Set2")
plt.title("Violin Plot of Tip Distribution Across Time (Lunch vs Dinner)")
plt.savefig('008.Violin_Plot_of_Tip_Distribution_Across_Time.png')
```
