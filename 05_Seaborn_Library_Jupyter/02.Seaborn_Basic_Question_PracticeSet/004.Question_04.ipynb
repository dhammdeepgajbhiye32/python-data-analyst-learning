```python
# Question 4: Create a boxplot showing the distribution of total_bill across different days.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Creating a boxplot to show the distribution of total_bill across different days.

sns.boxplot(data = tips , x = 'day', y = "total_bill", palette = 'Set2')
plt.title("Boxplot of Total Bill across Different Days")
plt.savefig('004.Boxplot_Total_Bill_across_Different_Days.png')
```
