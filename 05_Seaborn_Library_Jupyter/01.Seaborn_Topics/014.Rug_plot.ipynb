```python
# Here we will create rug plot using seaborn library.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
penguins = sns.load_dataset("penguins")
penguins

sns.rugplot(data = penguins,  y = 'body_mass_g',hue = "species",height = 0.5)
```
