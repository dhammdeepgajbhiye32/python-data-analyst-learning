```python
# Here we will learn how to create a pair plot in seaborn library
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sqlalchemy import column


penguins = sns.load_dataset('penguins')
penguins


columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
sns.pairplot(data = penguins, diag_kind = 'kde', hue = 'sex', palette = "Set2")
plt.suptitle("Pair Plot of Penguin Measurements by Sex", y=1.02)
plt.savefig("015.1.Pair_Plot.png")
```

```python
sns.pairplot(data = penguins, diag_kind = 'hist', hue = 'sex', palette = "Set2")
plt.suptitle("Pair Plot of Penguin Measurements by Sex (Histogram)", y=1.02)
plt.savefig("015.2.Pair_Plot_Histogram.png")
```
