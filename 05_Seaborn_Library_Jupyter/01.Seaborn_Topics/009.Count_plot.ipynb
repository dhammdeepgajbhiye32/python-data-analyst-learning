```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
penguins  = sns.load_dataset('penguins')
penguins


sns.countplot(data = penguins, x = 'species')
plt.title("Count Plot of Penguin Species")
plt.savefig("009.1.Count_Plot.png")
```

```python
sns.countplot(data = penguins, x = 'species', hue= 'island')
plt.title("Count Plot of Penguin Species by Island")
plt.savefig("009.2.Count_Plot.png")
```
