```python
# Here we will see how to create a heat map using seaborn library.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from encodings.punycode import T


penguins = sns.load_dataset("penguins")
penguins

columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
penguins[columns].corr()
sns.heatmap(data = penguins[columns].corr(), annot = True,cmap = 'Blues',vmin=-0.2,linewidths=0.5,linecolor='black')
```
