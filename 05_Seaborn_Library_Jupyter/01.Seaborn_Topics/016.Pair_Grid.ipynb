```python
# Here we will define a pair grid, which contain the upper triangle with kde, lower triangle with scatter and the diagonal with histogram
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from seaborn import PairGrid


penguins = sns.load_dataset('penguins')
penguins

graph = PairGrid(data = penguins, hue = 'sex', palette = "Set2")
graph.map_upper(sns.kdeplot)
graph.map_lower(sns.scatterplot)
graph.map_diag(sns.histplot)
graph.add_legend()
```
