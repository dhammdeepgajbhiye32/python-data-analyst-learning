```python
# Here we will see how to use line plot in seaborn library.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the penguins dataset 

penguins  = sns.load_dataset('penguins')
penguins
```

```python
# Here we will use the tips dataset for line plot. This dataset is available in seaborn library.
sns.set_style('whitegrid')
sns.lineplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island")
```
