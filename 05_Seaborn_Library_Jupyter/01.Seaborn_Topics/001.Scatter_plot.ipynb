```python
# importing libraries 

import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
# Loading the Penguins dataset into penguins variable

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# counting the values of species of the penguins dataset

penguins['species'].value_counts()
```

```python
# counting the values of island of the penguins dataset

penguins['island'].value_counts()
```

```python
# Creating the Scatter Plot to comapre the flipper lenth in mm vs body weight in grms.

sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g",hue="island")
plt.title("Scatter Plot of Flipper Length vs Body Mass")
plt.savefig("001.1.Scatter_plot.png")
```

```python
# Creating another Scatter plot to compare the flipper length in mm vs body mass in grms.

sns.set_style("ticks")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island",palette="Set2")
sns.despine()
plt.savefig("001.2.Scatter_plot.png")


# despine() is used to remove the top and right grids.
```

```python
# Trying to perform set.context() for each different types.
# 1. Paper
# 2. Notebook
# 3. Talk
# 4. Poster
```

```python
# 1. paper   - It is good for acadimic question paper size.

sns.set_style("ticks")
sns.set_context("paper")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island",palette="Dark2")
sns.despine()
plt.savefig("001.3.Scatter_plot.png")
```

```python
# 2. Notebook    - It is good for the Jupyter Notebook or Google Colab.

sns.set_style("ticks")
sns.set_context("notebook")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island",palette="Set1")
sns.despine()
plt.savefig("001.4.Scatter_plot.png")

```

```python
# 3. Talk     - It is good for Presentations.

sns.set_style("ticks")
sns.set_context("talk")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island",palette="Set1")
sns.despine()
plt.savefig("001.5.Scatter_plot.png")
```

```python
# 4. Poster      - It is good for big size posters.

sns.set_style("ticks")
sns.set_context("poster")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island",palette="Dark2")
sns.despine()
plt.savefig("001.6.Scatter_plot.png")
```
