```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
penguins = sns.load_dataset('penguins')
penguins

sns.set_style('whitegrid')
sns.boxplot(data = penguins , x = 'species', y = 'body_mass_g',hue= 'sex')
```
