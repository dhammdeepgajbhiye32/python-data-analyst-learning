```python
# Here vilon plot is same as box plot but it also shows the distribution of the data.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
penguins = sns.load_dataset('penguins')
penguins
sns.set_style('whitegrid')
sns.violinplot(data = penguins , x = 'species', y = 'body_mass_g',hue = "sex",split=True)
plt.title("Violin Plot of Body Mass by Penguin Species and Sex")
plt.savefig("011.1.Violin_Plot.png")
```
