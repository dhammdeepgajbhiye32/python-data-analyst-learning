```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np

penguins = sns.load_dataset('penguins')
penguins


# Create a joint plot with the penguins dataset 

sns.jointplot(data  = penguins , x = 'flipper_length_mm', y = 'body_mass_g')
```

```python
sns.jointplot(data  = penguins , x = 'flipper_length_mm', y = 'body_mass_g', kind = 'reg')
```
