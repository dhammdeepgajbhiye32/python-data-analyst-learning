```python
# Creating a Strip Plot to visualization for Species vs Body mass over Island
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Creating StirpPlots without jitter = True

sns.stripplot(data  = penguins, x = "species", y = "body_mass_g", hue = "island")
plt.title("Strip Plot of Species vs Body Mass")
plt.savefig("002.1.Strip_plot.png")
```

```python
# Creating StirpPlots with jitter = True

sns.stripplot(data  = penguins, x = "species", y = "flipper_length_mm", hue = "island", dodge = True)       # dodge is used to saperate the data points according to the hue.
plt.title("Strip Plot of Species vs Flipper Length")
plt.savefig("002.2.Strip_plot.png")
```
