```python
# Creating a bar plot with the penguins dataset. This dataset is available in seaborn library.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the penguins dataset 

penguins = sns.load_dataset('penguins')

penguins

# Creating a bar plot

sns.set_style('whitegrid')
sns.barplot(data = penguins, x = 'species', y = 'body_mass_g', hue="sex",palette=["lightblue", "pink"],estimator= np.sum)
plt.title("Bar Plot of Body Mass by Species and Sex")
plt.savefig("008.1.Bar_Plot.png")
```
