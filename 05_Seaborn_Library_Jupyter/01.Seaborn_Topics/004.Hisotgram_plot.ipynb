```python
# Here I am going to use Histogram to visualize the comparision of number of counts vs body mass in grams
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
# Creating Histogram

from numpy import stack


sns.histplot(data = penguins, x = "body_mass_g", hue = "sex", multiple= "stack")        # Here multiple is used to make  a single stack to compare both sex
```
