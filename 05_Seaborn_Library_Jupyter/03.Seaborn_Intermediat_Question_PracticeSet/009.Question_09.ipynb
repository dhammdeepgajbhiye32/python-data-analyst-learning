```python
# Question 9: Use catplot() to compare average tips by day grouped by smoker status.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# Creating a catplot to compare average tips by day grouped by smoker status
sns.catplot(data = tips, x = 'day', y = 'tip', hue = 'smoker', kind = 'bar')
plt.title('Average Tips by Day Grouped by Smoker Status')
plt.savefig('009.Catplot_Average_Tips_by_Day_Grouped_by_Smoker_Status.png')
```
