```python
# Question 2: Customize the color palette of a barplot using coolwarm.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load the tips dataset 

tips = sns.load_dataset('tips')
tips
```

```python
# Creating the barplot with customized color palette

sns.barplot(data = tips , x = 'day', y = 'total_bill', palette = "coolwarm")
plt.title("Barplot of Total Bill by Day with Coolwarm Palette")
plt.savefig('002.Barplot_of_Total_Bill_by_Day_with_Coolwarm_Palette.png')
```
