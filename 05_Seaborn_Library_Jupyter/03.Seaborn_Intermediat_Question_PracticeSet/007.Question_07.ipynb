```python
# Question 7: Overlay a boxplot and stripplot on the same graph.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# Creating a overlay of boxplot and stripplot to show tips distribution by day
plt.figure(figsize=(8, 6))
sns.boxplot(data = tips, x = 'day', y = 'tip', palette = 'Set2')
sns.stripplot(data = tips, x = 'day', y = 'tip', color = 'Blue', alpha = 0.5)
plt.title('Tips Distribution by Day (Boxplot + Stripplot)')
plt.savefig('007.Overlay_Boxplot_and_Stripplot_Tips_Distribution_by_Day.png')
```
