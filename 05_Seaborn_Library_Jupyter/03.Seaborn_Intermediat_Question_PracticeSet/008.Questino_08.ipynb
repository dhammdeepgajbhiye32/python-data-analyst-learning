```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset
tips  = sns.load_dataset('tips')
tips
```

```python
# Creating a distribution plot (hist + KDE) for the tip column

sns.displot(tips['tip'],kde = True)
plt.title('Distribution Plot for Tip Column')
plt.savefig("008.Distribution_Plot_for_Tip_Column.png")
```
