```python
# Question 5: Plot a line plot showing average total_bill by size (number of people).
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Creating a line plot to show average total_bill by size(number of people)

sns.lineplot(data = tips, x = 'size', y = 'total_bill', estimator = 'mean')
plt.title('Average Total Bill by Size')
plt.savefig('005.Line_Plot_Average_Total_Bill_by_Size.png')
```
