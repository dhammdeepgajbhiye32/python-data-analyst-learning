```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading the tips dataset
tips = sns.load_dataset('tips')
tips
```

```python
# using facietgrid to create separate scatter plots for lunch and dinner
g = sns.FacetGrid(data = tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.suptitle("Scatter Plots of Total Bill vs Tip for Lunch and Dinner", y=1.05)
plt.savefig('004.Scatter_Plots_of_Total_Bill_vs_Tip_for_Lunch_and_Dinner.png')  
```
