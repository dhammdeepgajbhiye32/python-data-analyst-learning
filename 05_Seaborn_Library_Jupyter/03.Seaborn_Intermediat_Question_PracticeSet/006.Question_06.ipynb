```python
# Question 6: Create a stripplot showing tips distribution by day.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the tips dataset

tips = sns.load_dataset('tips')
tips
```

```python
# Create a stripplot to show tips distribution by day

sns.stripplot(data  = tips, x = 'day', y = 'tip', palette = 'Set2')
plt.title('Tips Distribution by Day')
plt.savefig('006.Stripplot_Tips_Distribution_by_Day.png')
```
