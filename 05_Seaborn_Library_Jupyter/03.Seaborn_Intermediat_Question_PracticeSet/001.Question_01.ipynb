```python
# Question 1: Create a heatmap showing the correlation between numeric columns in the dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the tips dataset

penguins = sns.load_dataset('penguins')
penguins
```

```python
# Creating the heatmap of correlation between numeric columns

columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
penguins[columns].corr()
sns.heatmap(data = penguins[columns].corr(), annot=True, cmap='Blues')
plt.title("Heatmap of Correlation Between Numeric Columns")
plt.savefig('001.Heatmap_of_Correlation_Between_Numeric_Columns.png')
```
