```python
# Question 3: Create a scatter plot with hue = day and style = time.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load the tips dataset 

tips = sns.load_dataset('tips')
tips
```

```python
# Create as scatter plot with hue = day and style = time

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', style = 'time')
plt.title("Scatter Plot of Total Bill vs Tip with Hue = Day and Style = Time")
plt.savefig('003.Scatter_Plot_of_Total_Bill_vs_Tip_with_Hue_Day_and_Style_Time.png')
```
