## 📊 Exploratory Data Analysis

### 📈 Monthly Sales Trend - (Line chart)

```python
# Sum of Sales according to month

from pyparsing import line


Monthly_Sales = df.groupby("Month")['Sales'].sum()
Monthly_Sales = Monthly_Sales.reindex(["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
Monthly_Sales

```

```python
# Creating Visualizations  

# 1. Monthly Sales Trend Line Chart


plt.plot(Monthly_Sales.index, Monthly_Sales.values)
plt.title("Month wise Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.savefig("1_Month_wise_Sales.png")
plt.show()
```

### 🌍 Region-wise Sales - (Bar chart)

```python
# Calculating the Region wise Total Sales 

Region_Sales = df.groupby("Region")['Sales'].sum().apply(lambda x: format(x,",.1f"))
Region_Sales
```

```python
# Converting the Region Sales to float format from String

Region_Sales = Region_Sales.replace(',','', regex= True).astype(float)
```

```python
# Calcuating the Total Sales, Central Sales, and Central percent in Total Sales

Total_Sales = Region_Sales.sum()
central_sales = Region_Sales['Central']

central_Percent = (central_sales / Total_Sales) * 100
central_Percent = central_Percent.round(2)
central_Percent
```

### 📦 Top Products - (Vertical bar)

```python
# 2. Sales by Region - Bar Chart

from matplotlib.pyplot import bar


Region_Sales = df.groupby("Region")['Sales'].sum().sort_index().plot(kind= 'bar')
plt.title("Region Wise Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.savefig("2_Region_wise_Sales.png")
```

### 💰 Profit by Category - (Pie Chart)

```python
# 3. Profit Contribution by Category - Pie Chart

from enum import auto


Category_Profit = df.groupby("Category")['Profit'].sum().plot(kind='pie',autopct = "%1.1f%%")
plt.title("Category Wise Profit")
plt.ylabel('')
plt.savefig("3_Category_Wise_Profit.png")
```

### 💰 Profit vs Sales - (Scatter plot)

```python
# 4. Profit vs Sales Relationship - Scatter Plot

# Month Wise Profit

Profit_Month_wise = df.groupby('Month')['Profit'].sum()
Profit_Month_wise = Profit_Month_wise.reindex(["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
Profit_Month_wise
```

```python
# Product Wise Profit

Product_wise_Profit = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10).round(2)
Product_wise_Profit
```

```python
# Month Wise Sales

Monthly_Sales
```

```python
# Relationship of Profit and Sales over each month

plt.scatter(Monthly_Sales,Profit_Month_wise,color = "Orange")
plt.title("Profit over Sales Relationship")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid()
plt.savefig("4_Profit_over_Sales_Relationship.png")
plt.show()
```

### 💰 Sales Distribution - (Histogram plot)

```python
# Sales Distributions - Histogram

plt.hist(Monthly_Sales,bins = 20)
plt.title("Sales Distribution per Month")
plt.xlabel('Sales')
plt.savefig("5_Sale_Distribution_Per_Month.png")
plt.show()
```

### 💰 Top 10 Product by Sales - (Horizontal Bar plot)

```python
#  Top 10 Products by Sales - Horizontal Bar

from sqlalchemy import false


ax = (
    df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False).head(10).plot(kind='barh')
    
)
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product")
plt.savefig("6_Top_10_Products_by_Sales.png")
ax.invert_yaxis()
```

```python
# Creating year columnn

df['Year'] = df['Order Date'].dt.year
df
```

### 📈 Year-wise Growth - (Multi-Line Chart)

```python
#  Year-wise Growth - Multi-line chart

import calendar
from pandas import pivot_table

month_order = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

pivot_df = df.pivot_table(values="Sales",index="Month",columns="Year",aggfunc=sum)
pivot_df = pivot_df.reindex(month_order)
pivot_df.plot(marker = "o", figsize=(8,5))
plt.title("Sales Trend by Year")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid()
plt.savefig("7_Sales_trend_by_year.png")
plt.show()

```

### Region-wise Monthly Sales - (Stacked Bar)

```python
#  Region-wise Monthly Sales - Stacked Bar

month_order

pivot_df = df.pivot_table(values="Sales",index="Month",columns="Region",aggfunc=sum)
pivot_df = pivot_df.reindex(month_order)
pivot_df.plot(kind='bar',stacked= True,figsize=(10,10))
plt.title("Region wise Monthly Sales")
plt.ylabel("Sales")
plt.savefig("8_Region_wise_Monthly_Sales.png")
```

## Dashboard

```python
#  Dashboard Layout using subplots


fig, axs = plt.subplots(3,3, figsize=(20,15))
fig.suptitle("Sales Performance Analysis and Visualization",
             fontsize=16, fontweight="bold")


# -------------------- 1. Month wise Sales --------------------
axs[0,0].plot(Monthly_Sales.index, Monthly_Sales.values)
axs[0,0].set_title("Month wise Sales")
axs[0,0].set_xlabel("Month")
axs[0,0].set_ylabel("Sales")


# -------------------- 2. Region Wise Sales --------------------
df.groupby("Region")["Sales"].sum().sort_index()\
.plot(kind="bar", ax=axs[0,1])

axs[0,1].set_title("Region Wise Sales")
axs[0,1].set_xlabel("Region")
axs[0,1].set_ylabel("Sales")

# # -------------------- 3. Category Wise Profit (Pie) --------------------
# Category_Profit = df.groupby("Category")["Profit"].sum()

# axs[0,2].pie(
#     Category_Profit.values,
#     labels=Category_Profit.index,
#     autopct="%1.1f%%",
#     startangle=90
# )

# axs[0,2].set_title("Category Wise Profit")
# axs[0,2].set_ylabel("")

Category_Profit = df.groupby("Category")["Profit"].sum()

axs[0,2].pie(
    Category_Profit.values,
    labels = Category_Profit.index,
    autopct = "%1.1f%%",
    startangle = 90
    )

axs[0,2].set_title("Category Wise Profit")
axs[0,2].set_ylabel('')

# -------------------- 4. Profit vs Sales --------------------
axs[1,0].scatter(Monthly_Sales, Profit_Month_wise, color="orange")
axs[1,0].set_title("Profit over Sales Relationship")
axs[1,0].set_xlabel("Sales")
axs[1,0].set_ylabel("Profit")

# -------------------- 5. Histogram --------------------
axs[1,1].hist(Monthly_Sales, bins=20)
axs[1,1].set_title("Sales Distribution per Month")
axs[1,1].set_xlabel("Sales")

# -------------------- 6. Top 10 Products --------------------
df.groupby("Product Name")["Sales"].sum()\
.sort_values(ascending=False).head(10)\
.plot(kind="barh", ax=axs[1,2])

axs[1,2].set_title("Top 10 Products by Sales")
axs[1,2].set_xlabel("Sales")
axs[1,2].set_ylabel("Product")
axs[1,2].invert_yaxis()


# -------------------- 7. Sales Trend by Year --------------------
pivot_year = df.pivot_table(values="Sales", index="Month",
                            columns="Year", aggfunc="sum")

pivot_year = pivot_year.reindex(month_order)

pivot_year.plot(marker="o", ax=axs[2,0])

axs[2,0].set_title("Sales Trend by Year")
axs[2,0].set_xlabel("Months")
axs[2,0].set_ylabel("Sales")

# -------------------- 8. Region wise Monthly Sales --------------------
pivot_region = df.pivot_table(values="Sales", index="Month",
                              columns="Region", aggfunc="sum")

pivot_region = pivot_region.reindex(month_order)

pivot_region.plot(kind="bar", stacked=True, ax=axs[2,1])

axs[2,1].set_title("Region wise Monthly Sales")
axs[2,1].set_ylabel("Sales")


# -------------------- 9. Hide empty subplot --------------------
axs[2,2].axis("off")



plt.tight_layout()

plt.savefig("9.Dashboard.png")

plt.show()

```
