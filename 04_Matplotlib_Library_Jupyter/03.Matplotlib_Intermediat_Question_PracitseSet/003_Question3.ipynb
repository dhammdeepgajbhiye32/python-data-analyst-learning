```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/dhamm/Downloads/company_monthly_sales_dataset.csv")
df
```

```python
# Groupping by Category and Calculating Sales 

Groupby_Categroy = df.groupby("Category")["Sales"].sum()
Groupby_Categroy
```

```python
# Creating Bar Chart 

plt.bar(Groupby_Categroy.index,Groupby_Categroy.values)
plt.title("Category wise Total Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales in Million")
plt.savefig("003_Category wise Total Sales.png")
plt.show()
```
