```python
# Create subplots (2x2) with line, bar, scatter, and histogram in matplotlib to visualize the salary vs age.
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Creating a dataset of Employee names, Salary and Age

data = {
    "Name" : ["Shivam","Rudransh","Bhumika","Priya","Shivi","Ritika"],
    "Salary": [25000,28000,54000,34000,24790,34800],
    "Age" : [25,27,30,32,26,31],
    "Department" : ["IT","HR","Finance","Finance","HR","HR"]
}

df = pd.DataFrame(data)
df
```

```python
# Creating subplots for showing line, bar, scatter, and histogram plots in single Plot

from contourpy import LineType
from matplotlib import markers


fig, axs = plt.subplots(1,4, figsize = (15,5))

# Sorting the age column

sort_Age = df.sort_values("Age")
sort_Age

# Line plot

axs[0].plot(sort_Age["Age"],sort_Age["Salary"])
axs[0].set_title("Age vs Salary of Employee")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

# count value

count_value = df["Department"].value_counts()
count_value

# Bar plot

axs[1].bar(count_value.index,count_value.values)
axs[1].set_title("Age vs Salary of Employee")
axs[1].set_xlabel("Age")
axs[1].set_ylabel("Salary")

# Scatter plot

axs[2].scatter(df["Salary"],df["Age"])
axs[2].set_title("Age vs Salary of Employee")
axs[2].set_xlabel("Age")
axs[2].set_ylabel("Salary")

# Histogram plot

axs[3].hist(df["Salary"],bins = 15)
axs[3].set_title("Age vs Salary of Employee")
axs[3].set_xlabel("Age")
axs[3].set_ylabel("Salary")

plt.savefig("001_Multiple_img.png")

# Printing the sub-plots within plot

plt.show()
```
