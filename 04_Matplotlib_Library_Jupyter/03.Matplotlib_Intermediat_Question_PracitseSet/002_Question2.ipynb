```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df = pd.read_csv("C:/Users/dhamm/Downloads/company_monthly_sales_dataset.csv")
df
```

```python
df = df.query('Company == "Company_1"')[["Company","Month","Region","Category","Sales"]]
df
```

```python
# Plotting the line chart for company 1 

plt.plot(df["Sales"],df["Month"],marker = "*",color = "Blue")
plt.title("Month wise Sales of Company 1")
plt.xlabel("Sales")
plt.ylabel("Month")
plt.savefig("002.Month wise Sales of Company 1.png")
plt.show()
```

```python
# Groupping the Sales by Months

Groupby_Month = df.groupby("Month")['Sales'].sum().sort_values()
Groupby_Month
```

```python
# Plotting the line chart

plt.plot(Groupby_Month.index,Groupby_Month.values,marker = "*")
plt.title("Month wise Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.savefig("002_Month wise Sales.png")
plt.show()
```
