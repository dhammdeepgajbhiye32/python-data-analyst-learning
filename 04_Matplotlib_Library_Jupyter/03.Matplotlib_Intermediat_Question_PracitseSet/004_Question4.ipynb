```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/dhamm/Downloads/company_monthly_sales_dataset.csv")
df
```

```python
# Groupping the Sales by Category

groupbyCategory = df.groupby("Category")['Sales'].sum()
groupbyCategory
```

```python
# Plotting the pie chart 

count = df['Category'].value_counts()
count
```

```python
plt.pie(count,labels = groupbyCategory.index,autopct = "%1.2f%%",explode=[0.1,0.1,0.1],shadow= True)
plt.title("Category wise percent Share")
plt.savefig("004_Category wise Percentage Share.")
plt.show()
```
