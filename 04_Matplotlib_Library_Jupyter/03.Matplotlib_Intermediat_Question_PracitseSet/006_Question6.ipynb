```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# data set 

df = pd.read_csv("C:/Users/dhamm/Downloads/company_monthly_sales_dataset.csv")
df
```

```python
# Groupping the Sales with respect to month.

GrouppingSales = df.groupby("Month")["Sales"].sum().sort_values()
GrouppingSales
```

```python
# plotting the line plot 

from debugpy import Endpoint

plt.plot(GrouppingSales.values, GrouppingSales.index,color = "Blue")
plt.scatter(GrouppingSales.values[0],GrouppingSales.index[0], marker = "*",color = "Red")
plt.scatter(GrouppingSales.values[-1],GrouppingSales.index[-1], marker = "*",color = "Red")
plt.title("Month wise Sales")
plt.xlabel("Sales")
plt.ylabel("Months")
plt.savefig("006.MonthWiseSales.png")
plt.show()
```
