```python
# Create Scatter plot showing height vs weight.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

height = np.random.randint(160,200,12)
print(height)
weight = np.random.randint(60,90,12)
print(weight)

# plotting the Scatter plot 

plt.scatter(weight,height, color = "SkyBlue",marker= "o")
plt.title("Scatter Plot for Height vs Weight")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.savefig("005_Scatter plot showing height vs weight.png")
plt.show()
```
