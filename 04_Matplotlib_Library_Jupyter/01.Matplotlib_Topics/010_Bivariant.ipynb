```python
# Here we will fetch the salary of HR department
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = {
    "Salary" : [25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 18000, 26000, 26000, 45000,32450],
    "Department" : ["HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR"],
    "Age" : [25, 30, 35, 29, 37, 22, 29, 32, 40, 45, 19, 28, 26, 31, 38, 27]
}
df = pd.DataFrame(data)
df
```

```python
Hr_sal = df[df['Department'] =='HR']['Salary']
Hr_sal
```

```python
IT_sal = df[df['Department'] =='IT']['Salary']
IT_sal
```

```python
Finance_sal = df[df['Department'] =='Finance']['Salary']
Finance_sal
```

```python
# Creatiing a BOX plot for the salary of HR, IT and Finance department
plt.boxplot([Hr_sal, IT_sal, Finance_sal], labels=['HR', 'IT', 'Finance'])
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.savefig("010_salary_boxplot.png")
plt.show()
```

```python
# another method is to use pie chart to show the distribution of salary by department

salary_by_departmemt = df.groupby("Department")['Salary'].sum()
salary_by_departmemt
```

```python
# Creating pie chart to show the distribution of salary by department

plt.pie(salary_by_departmemt, labels = salary_by_departmemt.index, autopct = "%1.1f%%",explode = [0.1, 0.1, 0.1],shadow=True)
plt.title("Average Salary Distribution by Department")
plt.savefig("010_salary_pie_chart.png")
plt.show()
```

```python
# showing the mean salary by department using bar chart

from matplotlib.patches import Shadow

mean_salary = df.groupby("Department")['Salary'].mean()
plt.bar(mean_salary.index, mean_salary.values, color=['blue', 'orange', 'green'])
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.grid()
plt.savefig("010_salary_bar_chart.png")
plt.show()
```
