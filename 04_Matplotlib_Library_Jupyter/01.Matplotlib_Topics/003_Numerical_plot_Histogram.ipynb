```python
# Creating a Histogram plot using Matplotlib
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "Salary":[25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 78000, 90000, 26000, 45000]
}
df = pd.DataFrame(data)
df.head(5)
```

```python
# Plotting the Salary column in a histogram

plt.hist(df["Salary"], bins=15, edgecolor='green')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("No. of Employees")
plt.savefig("003_salary_histogram.png")
plt.show()
```
