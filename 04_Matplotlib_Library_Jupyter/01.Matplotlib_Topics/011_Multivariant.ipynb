```python
# Here we will see how to create multivariate plots using Matplotlib.import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "Salary" : [25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 18000, 26000, 26000, 45000,32450],
    "Department" : ["HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR"],
    "Age" : [25, 30, 35, 29, 37, 22, 29, 32, 40, 45, 19, 28, 26, 31, 38, 27],
    "Experience" : [2, 5, 10, 3, 8, 1, 4, 6, 12, 15, 0.5, 2.5, 3.5, 7, 9, 4]
}
df = pd.DataFrame(data)
df
```

```python
# Showing the relationship betwenn Salary, Age and Experience using a Bubble plot.


plt.scatter(df["Age"], df["Salary"], s = df["Experience"]*20, color = "SkyBlue", edgecolors= "Black")
plt.title("Salary vs Age with Experience as size")
plt.xlabel("Age") 
plt.ylabel("Salary") 
plt.savefig("011_Multivariant_bubble_chart.png")
plt.show()
```
