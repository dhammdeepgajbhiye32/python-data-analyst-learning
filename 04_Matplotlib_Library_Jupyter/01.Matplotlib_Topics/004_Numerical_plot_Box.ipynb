```python
# Here we are creating a box plot using Matplotlib to visualize the distribution of the Salary column in our DataFrame.

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "Salary":[25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 78000, 90000, 26000, 45000,32450]
}
df = pd.DataFrame(data)
df
```

```python
# Plotting the Salary column
plt.boxplot(df["Salary"])
plt.grid()
plt.savefig("004_Numerical_plot_Box.png")
plt.show()
```
