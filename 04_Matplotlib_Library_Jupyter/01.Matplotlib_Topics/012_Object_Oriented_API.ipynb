```python
# Creating Object Oriented API for Matplotlib. This is a more advanced way to create plots, giving you more control over the plot elements and allowing you to 
# reate more complex visualizations.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
data = {
    "Salary" : [25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 18000, 26000, 26000, 45000,32450],
    "Department" : ["HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR"],
    "Age" : [25, 30, 35, 29, 37, 22, 29, 32, 40, 45, 19, 28, 26, 31, 38, 27],
    "Experience" : [2, 5, 10, 3, 8, 1, 4, 6, 12, 15, 0.5, 2.5, 3.5, 7, 9, 4]
}

df = pd.DataFrame(data)
df
```

```python
# Line Plot

sort_Age = df.sort_values('Age')
sort_Age
```

```python
fig, axs = plt.subplots(1,3, figsize  = (20,5))

# line plot

axs[0].plot(sort_Age['Age'], sort_Age['Salary'])
axs[0].set_title('Line Plot')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Salary')

# Bar Plot

count_dept = df['Department'].value_counts()
count_dept

axs[1].bar(count_dept.index, count_dept.values)
axs[1].set_title('Bar Plot')
axs[1].set_xlabel('Department') 
axs[1].set_ylabel('Count')

# Box Plot

axs[2].boxplot(df['Salary'])
axs[2].set_title('Box Plot') 
axs[2].set_ylabel('Salary') 

plt.savefig('012_object_oriented_api.png') # Save the figure as a file

plt.show()
```
