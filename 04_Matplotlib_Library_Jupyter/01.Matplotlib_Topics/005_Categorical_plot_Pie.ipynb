```python
# Creating a pie chart to visualize the distribution of employees across different departments
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a sample dataset

data = {
    "Salary":[25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 78000, 90000, 26000, 45000,32450]
}
df = pd.DataFrame(data)
df
```

```python
# Adding a Department column to the DataFrame for pie chart visualization
df['Department'] = ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance', 'Finance', 'HR', 'IT', 'Finance','HR', 'IT', 'HR', 'IT', 'Finance', 'HR']
df
```

```python
# Creating a pie chart

count = df['Department'].value_counts()
count
```

```python
plt.pie(count,labels = count.index, autopct = "%1.1f%%", explode = [0.1,0.1,0.1],shadow = True)
plt.title("Department Distribution")
plt.savefig("005_department_pie_chart.png")
plt.show()
```
