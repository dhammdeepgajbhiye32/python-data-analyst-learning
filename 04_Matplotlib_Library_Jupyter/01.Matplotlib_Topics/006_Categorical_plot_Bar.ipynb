```python
# Creating a Bar chart to visualize the count of employees in each department

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "Salary":[25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 78000, 90000, 26000, 45000,32450]
}
df = pd.DataFrame(data)
df
```

```python
# Adding a Department column to the DataFrame for pie chart visualization
df['Department'] = ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance', 'Finance', 'HR', 'IT', 'Finance','HR', 'IT', 'HR', 'IT', 'Finance', 'HR']
df
```

```python
# Counting the number of employees in each department
count = df['Department'].value_counts()
count
```

```python
# Plotting the Bar chart for department distribution
plt.bar(count.index,count.values, color=['blue', 'orange', 'green'])
plt.title("Department Distribution")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.savefig("006_department_bar_chart.png")
plt.show()
```
