```python
# Creating a numerical plot using Matplotlib and Pandas with column name is "Salary"
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  
```

```python
data = {
    "Salary":[25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 78000, 90000, 26000, 45000]
}
df = pd.DataFrame(data)
df.head(5)
```

```python
# Total number of rows in the dataset
df.shape
```

```python
# Plotting the Salary column using line plot

plt.plot(df["Salary"],color = "red",marker = "o",linestyle = "-")
plt.title("Salary Distribution")
plt.xlabel("Index")
plt.ylabel("Salary")
plt.grid()
plt.savefig('002_salary_line_plot.png') # Save the plot as a PNG file
plt.show()
```
