```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np

# Create a DataFrame with two numerical columns and a categorical column
data = {
    "Salary" : [25000, 30000, 35000, 29000, 37000, 12000, 29000, 22000, 42000, 45000, 19000, 18000, 26000, 26000, 45000,32450],
    "Deparment" : ["HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR"],
    "Age" : [25, 30, 35, 29, 37, 22, 29, 32, 40, 45, 19, 28, 26, 31, 38, 27]
}

df = pd.DataFrame(data)
df
```

```python
# Scatter plot of Salary vs Age
from numpy import size


plt.scatter(df["Age"], df["Salary"], color='Orange')
plt.title("Salary vs Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid()
plt.savefig("007_salary_age_scatter.png")
plt.show()
```
