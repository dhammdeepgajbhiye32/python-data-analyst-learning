```python
import matplotlib.pyplot as plt
import pandas as pd


# Create a simple line plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title('Basic Line Plot')
plt.xlabel('Price')
plt.ylabel('Sales')
plt.grid()
plt.savefig('001_line_plot.png') # Save the plot as a PNG file
plt.show()
```
