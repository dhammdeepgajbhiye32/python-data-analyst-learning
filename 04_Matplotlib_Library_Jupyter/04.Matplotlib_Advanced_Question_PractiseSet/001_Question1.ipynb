```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/dhamm/Downloads/company_monthly_sales_dataset.csv")
df
```

```python
# Creating the line plot for 3 different companies

# Company 1

Company1 = df.query('Company =="Company_1"')[['Company','Month','Region','Sales']]
Company1
```

```python
# Company 2

Company2 = df.query('Company =="Company_2"')[['Company','Month','Region','Sales']]
Company2
```

```python
# Company 3

Company3 = df.query('Company =="Company_3"')[['Company','Month','Region','Sales']]
Company3
```

```python
from matplotlib import legend


plt.plot(Company1['Month'], Company1['Sales'], label='Company 1')
plt.plot(Company2['Month'], Company2['Sales'], label='Company 2')
plt.plot(Company3['Month'], Company3['Sales'], label='Company 3')
plt.title("3 Different Company Monthly Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.savefig("001.3 Different Company Monthly Sales Comparison.png")
plt.show()
```
