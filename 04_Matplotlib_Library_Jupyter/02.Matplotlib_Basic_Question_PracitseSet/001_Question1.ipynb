```python
# Plot a simple line chart of numbers from 1 to 10
```

```python
import matplotlib.pyplot as plt

x = list(range(1, 11))
y = x

plt.plot(x, y,marker = '*')
plt.title("Line Chart of Numbers from 1 to 10")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.savefig("001_line_chart.png")
plt.show()
```
