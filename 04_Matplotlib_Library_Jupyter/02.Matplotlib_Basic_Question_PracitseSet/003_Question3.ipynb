```python
import matplotlib.pyplot as plt

x = list(range(1, 11))
y = [x**2 for x in x] # Example data: y = x^2

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Line Graph of x vs y')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.savefig("003_line_graph.png") # Save the figure as a file
plt.show()
```
