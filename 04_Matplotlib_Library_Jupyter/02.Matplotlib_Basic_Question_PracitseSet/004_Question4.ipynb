```python
# Change line color, style and width in a line graph.
import matplotlib.pyplot as plt
x = list(range(1, 11))
y = [x**2 for x in x] # Example data: y = x^2

plt.plot(x,y, marker = '*', color = 'Green', linestyle = '--', linewidth = 2)
plt.title('Line Graph of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.savefig('004_line_graph.png')
plt.show()
```
