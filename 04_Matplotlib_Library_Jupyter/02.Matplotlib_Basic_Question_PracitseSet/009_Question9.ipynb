```python
# plot a histogram of 100 random numbers using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np
# Generate random data
x = np.random.randint(1,101,100)

# Plot histogram
plt.hist(x, bins=10, color='blue', edgecolor='black')
plt.title('Histogram of 100 Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.savefig('009_Histogram.png')
plt.show()
```
