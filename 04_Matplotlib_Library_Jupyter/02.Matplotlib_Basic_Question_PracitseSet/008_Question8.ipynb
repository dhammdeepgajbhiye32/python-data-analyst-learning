```python
# Create a Histogram bar chart using Random numbers in Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data

x = np.random.randint(1,100,50) # 50 random integers between 1 and 100

plt.hist(x, bins=10, color='blue', edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.savefig('008_Histogram.png')
plt.show()
```
