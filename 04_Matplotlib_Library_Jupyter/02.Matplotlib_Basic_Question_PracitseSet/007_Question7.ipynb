```python
# Plot a Bar chart of 5 students marks in a subject using Matplotlib.
```

```python
import matplotlib.pyplot as plt
marks = [85, 90, 78, 92, 88]
subject = 'Math'
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
plt.bar(students, marks, color=['blue', 'green', 'red', 'orange', 'purple'])
plt.title(f'Bar Chart of {subject} Marks')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.savefig('007_Bar_Chart.png')
plt.show()
```
