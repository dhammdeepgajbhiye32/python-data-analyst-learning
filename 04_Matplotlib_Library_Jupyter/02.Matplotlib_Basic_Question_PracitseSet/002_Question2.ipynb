```python
# Plot a line graph of x = [1,2,3,4,5] and y = [10,20,25,30,40]
```

```python
import matplotlib.pyplot as plt


x = list(range(1,6))
y = [10,20,25,30,40] 

plt.plot(x,y, marker = '*')
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.savefig("002_line_graph.png")
plt.show()
```
