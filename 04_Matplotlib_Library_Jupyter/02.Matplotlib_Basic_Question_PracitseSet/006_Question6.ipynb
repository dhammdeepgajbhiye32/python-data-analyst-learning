```python
# Create a scatter plot using Random numbers in Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.randint(1,50, 10) # 10 random integers between 1 and 50
y = np.random.randint(1,50, 10) # 10 random integers between 1 and 50   

plt.scatter(x, y, color='r', marker='o')
plt.title('Scatter Plot of Random Numbers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.savefig('006_scatter_plot.png')
plt.show()
```
